```python
import pandas as pd
movies_df = pd.read_csv("movies.csv")
movies_df
```

### Qual o diretor dirigiu mais filmes?

```python
Diretor = movies_df.sort_values("Director")[["Director", "Title"]]
Contador = Diretor.groupby("Director").count()
Maior = Contador.sort_values("Title", ascending=False)
print("O diretor que dirigiu mais filmes foi:")
Maior.iloc[0:1]
```

### Quais filmes tiverem os maiores arrecadações (Gross) em relação ao orçamento de produção (Budget)?

```python
# ADICIONE O SEU CÓDIGO AQUI
Gross_Budget = movies_df.sort_values("Title")[["Title", "Worldwide Gross", "Production Budget"]]
Gross_Budget["Lucro"] = Gross_Budget["Worldwide Gross"] - Gross_Budget["Production Budget"]
Maior_Lucro = Gross_Budget.sort_values("Lucro", ascending=False)
print("Os 10 filmes que tiveram maiores arrecadações em relação ao orçamento de produção, foi:")
Maior_Lucro.iloc[0:10]
```
